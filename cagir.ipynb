```python
import pandas as pd
import joblib

# Nesneleri yükle
model = joblib.load('mpg_poly_model.pkl')
poly = joblib.load('poly_transformer.pkl')
scaler = joblib.load('mpg_scaler.pkl')
expected_features = joblib.load('feature_names.pkl')

# Yeni veri örneği (Sözlük yapısı sütun sırasını korumak için en güvenlisidir)
yeni_veri_dict = {
    'weight': [3000],
    'acceleration': [15.5],
    'model_year': [80],
    'origin_japan': [0],
    'origin_usa': [1]
}
yeni_df = pd.DataFrame(yeni_veri_dict)

# Sütunların sırasının eğitimle aynı olduğundan emin ol
yeni_df = yeni_df[expected_features]

# DÖNÜŞTÜRME SIRASI:
# 1. Polynomial
yeni_poly = poly.transform(yeni_df)
# 2. Scale
yeni_scaled = scaler.transform(yeni_poly)
# 3. Predict
sonuc = model.predict(yeni_scaled)

print(f"Tahmin Edilen MPG: {sonuc[0]:.2f}")
```

```python
import joblib
import pandas as pd

# Kaydettiğin dosyaları yükle
model = joblib.load('mpg_poly_model.pkl')
poly = joblib.load('poly_transformer.pkl')
scaler = joblib.load('mpg_scaler.pkl')

print("Model ve dönüştürücüler yüklendi!")
```

```python
# Örnek: 1982 model, 3000 kg, 15 saniye hızlanma süresi olan bir ABD arabası
yeni_araba = {
    'weight': [3000],
    'acceleration': [15.0],
    'model_year': [82],
    'origin_japan': [0],
    'origin_usa': [1]
}

# Veriyi DataFrame haline getir (Sütun isimleri eğitimle aynı olmalı)
yeni_df = pd.DataFrame(yeni_araba)

# ADIM 1: Polinom Özelliklerine Dönüştür
# Bu işlem 5 olan özellik sayısını modelin beklediği sayıya (örn: 20) çıkarır
yeni_poly = poly.transform(yeni_df)

# ADIM 2: Veriyi Ölçeklendir (Scaling)
# Modelin "milyonluk" hatalar yapmaması için bu adım şarttır
yeni_scaled = scaler.transform(yeni_poly)

# ADIM 3: Tahmin Et
tahmin = model.predict(yeni_scaled)

print(f"Tahmin Edilen Yakıt Tüketimi (MPG): {tahmin[0]:.2f}")
```
